```python
print(f'Size of Data Frame\n{df.shape}\n\n')
print(f'Main characteristics and types of datas in Data Frame\n\n{df.describe()}\n\n')
print(df.info())
```

```python
df_procent_discount = df.assign(Procent_discount = df['DISC'] / df['Profit Margin'] * 100)
data_analisys_of_discount = pd.pivot_table(df, index=['City'], values=['DISC', 'Profit Margin'], aggfunc=[np.sum])

data_analisys_of_procent_discount = pd.pivot_table(df_procent_discount, index=['City'], values=['Procent_discount'], aggfunc=[np.mean])
print(data_analisys_of_discount.join(data_analisys_of_procent_discount, on=None, how='left'))

fig = plt.figure(figsize=(25, 15))
data_analisys_of_discount.plot(rot=30, stacked=True, kind='bar', xlabel='City', ylabel='EU', title='The volume of discounts provided by stores, taking into account the profit received')
plt.show();
print('Згідно проведеного аналізу данних по дисконту магазинів на продаж товару - бачимо незначну маржу у магазині "Woodlands", меньшу ніж у інших магазинів в 1,5-2 рази. Середній дисконт цього магазину - 4,2949. Вважаю необхідним зробити додаткову аналітику з продажу групп товарів та розглянути заходи збільшення загального середнього % дисконта до відмітки 4,4262')
```

```python
df_sales_income = df.assign(Sales_income = df['Operating Cost'] + df['Profit Margin'] + df['Product Cost'], Operating_costs_to_volume_of_goods = df['Operating Cost'] / df['QTY'])
data_analisys_of_Operating_Cost = pd.pivot_table(df_sales_income, index=['City'], values=['Operating_costs_to_volume_of_goods', 'QTY', 'Operating Cost', 'Profit Margin', 'Sales_income'], aggfunc = {'Operating_costs_to_volume_of_goods': [np.mean], 'QTY' : [np.sum], 'Operating Cost' : [np.sum], 'Profit Margin' : [np.sum], 'Sales_income' : [np.sum]})
print(f'{data_analisys_of_Operating_Cost}\n')
print("Операційні расходи становлять 15% від доходів магазинів, наряду з цим об'єм продажу товару в кожному магазині свій. Треба доповнити данні в основній таблиці більш детальною інформацією по операційних расходах, таких як расходи на ЗП сотрудників магазину, расходи на єнергетику, аренду приміщень, складської та торгівельної площі, або/та аренду земельної ділянки, расходи на логістику та інше. Це сприяє вивченню затрат, які знижуют маржу магазинів. Також для кожного магазину і продавця потрібно ввести мотиваційни коєфіціенти, які будуть залежати від об'ему продажі 'Operating_costs_to_volume_of_goods', та маржі - що буде знижати вплив (процент) операційних расходів на єкономічну діяльність магазинів")
```

```python
df_Male_subcategory = df[df.Gender == "Male"]['Sub_Category'].value_counts()
display('Перші десять підкатегорій товарів за кількістю закупівель чоловиками')
display(df_Male_subcategory.head(10))
plt.figure(figsize=(20, 10)) 

plt.subplot(1, 2, 1)  
plt.pie(df_Male_subcategory.head(10), labels=df_Male_subcategory.head(10).index, autopct='%1.0f%%')

plt.subplot(1, 2, 2)
plt.bar(df_Male_subcategory.head(10).index, df_Male_subcategory.head(10), )
plt.xlabel('Subcategory', fontsize=12)
plt.ylabel('Count of first 10 goods', fontsize=12)
plt.xticks(rotation = 90)



plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)  
plt.pie(df_Male_subcategory.tail(10), labels=df_Male_subcategory.tail(10).index, autopct='%1.0f%%')

plt.subplot(2, 2, 2)
plt.bar(df_Male_subcategory.tail(10).index, df_Male_subcategory.tail(10), )
plt.xlabel('Subcategory', fontsize=12)
plt.ylabel('Count of last 10 goods', fontsize=12)
plt.xticks(rotation = 90)

display('Останні десять підкатегорій товарів за кількістю закупівель чоловиками')
display(df_Male_subcategory.tail(10))

plt.tight_layout();

print('Концентрація маркетингових програм для чоловиків на перших 10 підкатегоріях для збільшення доходу магазинів. Розподілення загрузки персонала магазинів для концентрації і збільшення попиту на вказані групи.')
```

```python
df_Female_subcategory = df[df.Gender == "Female"]['Sub_Category'].value_counts()
display('Перші десять підкатегорій товарів за кількістю закупівель жінками')
display(df_Female_subcategory.head(10))
plt.figure(figsize=(20, 10)) 

plt.subplot(1, 2, 1)  
plt.pie(df_Female_subcategory.head(10), labels=df_Female_subcategory.head(10).index, autopct='%1.0f%%')

plt.subplot(1, 2, 2)
plt.bar(df_Female_subcategory.head(10).index, df_Female_subcategory.head(10), )
plt.xlabel('Subcategory', fontsize=12)
plt.ylabel('Count of first 10 goods', fontsize=12)
plt.xticks(rotation = 90)



plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)  
plt.pie(df_Female_subcategory.tail(10), labels=df_Female_subcategory.tail(10).index, autopct='%1.0f%%')

plt.subplot(2, 2, 2)
plt.bar(df_Female_subcategory.tail(10).index, df_Female_subcategory.tail(10), )
plt.xlabel('Subcategory', fontsize=12)
plt.ylabel('Count of last 10 goods', fontsize=12)
plt.xticks(rotation = 90)

display('Останні десять підкатегорій товарів за кількістю закупівель жінками')
display(df_Female_subcategory.tail(10))

plt.tight_layout();

print('Концентрація маркетингових програм для жінок на перших 10 підкатегоріях для збільшення доходу магазинів. Розподілення загрузки персонала магазинів для концентрації і збільшення попиту на вказані групи. Збільшення номенклатури товару у ціх підгрупах')
```
